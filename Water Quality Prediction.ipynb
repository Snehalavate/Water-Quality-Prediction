# Partitioning

```python
X = df.drop('Potability',axis=1)
```

```python
Y= df['Potability']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size= 0.2, random_state=101,shuffle=True)
```

```python
Y_train.value_counts()
```

```python
Y_test.value_counts()
```

# DT

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,confusion_matrix
dt=DecisionTreeClassifier(criterion= 'entropy', min_samples_split= 10, splitter= 'best')
dt.fit(X_train,Y_train)
```

```python
prediction=dt.predict(X_test)
accuracy_dt=accuracy_score(Y_test,prediction)*100
accuracy_dt
```

```python
print("Accuracy on training set: {:.3f}".format(dt.score(X_train, Y_train)))
print("Accuracy on test set: {:.3f}".format(dt.score(X_test, Y_test)))
```

```python
accuracy_score(prediction,Y_test)
```

```python
dt_cm=confusion_matrix(prediction,Y_test)
dt_cm
```

```python
import seaborn as sns
plt.title('Decision Tree')
sns.heatmap(dt_cm,annot=True)
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(metric='manhattan', n_neighbors=151)
knn.fit(X_train,Y_train)
```

```python
prediction_knn=knn.predict(X_test)
accuracy_knn=accuracy_score(Y_test,prediction_knn)*100
print('accuracy_score score     : ',accuracy_score(Y_test,prediction_knn)*100,'%')
```

```python
knn_cm=confusion_matrix(prediction_knn,Y_test)
knn_cm
```

```python
plt.title('KNN')
sns.heatmap(knn_cm,annot=True)
```

```python
#logistic Regression

from sklearn.linear_model import LogisticRegression
from sklearn import metrics
lr=LogisticRegression()
lr.fit(X_train,Y_train)
Y_pred_lr=lr.predict(X_test)
accuracy_lr=accuracy_score(Y_test,Y_pred_lr)*100
print(accuracy_lr)
```

```python
lr_cm=confusion_matrix(Y_pred_lr,Y_test)
lr_cm
```

```python
plt.title('Logistic Regression')
sns.heatmap(lr_cm,annot=True)
```

```python
from sklearn.svm import SVC
classifier=SVC(kernel='rbf',random_state=1)
classifier.fit(X_train,Y_train)
Y_pred_svc=classifier.predict(X_test)
accuracy_svc=accuracy_score(Y_test,Y_pred_svc)*100
print(accuracy_svc)
```

```python
svc_cm=confusion_matrix(Y_pred_svc,Y_test)
svc_cm
```

```python
plt.title('Support Vector Classifier')
sns.heatmap(svc_cm,annot=True)
```
